```python
import requests
#from google.colab import userdata

#API_TOKEN = userdata.get('ACCESS_TOKEN')
#BASE_URL = "https://api.github.com/repos/DataTalksClub/data-engineering-zoomcamp/events"
base_url = "https://api.github.com/repos/DataTalksClub/data-engineering-zoomcamp/events"

def paginated_getter():
    page_number = 1
    url = base_url

    # Iterate over pages
    while True:
        # Starts at **page 1** and makes a **GET request** to the API.
        #response = requests.get(url, headers={"Authorization": f"Bearer {API_TOKEN}"})
        response = requests.get(url)
        # Retrieves **JSON data**.
        page_data = response.json()
        yield page_data

        # Looks for the **"next" page URL** in the response headers.
        next_page = response.links.get("next", {}).get("url")


        # If there’s **no next page**, stops fetching and ends the loop.
        if not next_page:
            break

        # If a **next page exists**, updates `BASE_URL` and requests more data.
        page_number += 1
        url = next_page


for page_data in paginated_getter():
    print(page_data)
```

```python
#Using DLT
import dlt
from dlt.sources.helpers.rest_client import RESTClient
#from dlt.sources.helpers.rest_client.auth import BearerTokenAuth
from dlt.sources.helpers.rest_client.paginators import HeaderLinkPaginator
#from google.colab import userdata

#API_TOKEN = userdata.get('ACCESS_TOKEN')


def paginated_getter():
    client = RESTClient(
        base_url = "https://api.github.com/repos/DataTalksClub/data-engineering-zoomcamp/events",
        #auth=BearerTokenAuth(token=API_TOKEN),
        paginator=HeaderLinkPaginator(links_next_key="next"),
    )

    for page in client.paginate(""):    #if endpoint isnt in the base url add it here
        yield page


for page_data in paginated_getter():
    print(page_data)
```

```python
#setting up a dlt resource
@dlt.resource()
def paginated_getter():
    client = RESTClient(
        base_url = "https://api.github.com/repos/DataTalksClub/data-engineering-zoomcamp/events",
        #auth=BearerTokenAuth(token=API_TOKEN),
        paginator=HeaderLinkPaginator(links_next_key="next"),
    )

    for page in client.paginate(""):    #if endpoint isnt in the base url add it here
        yield page

for record in paginated_getter():
    print(record)
```

```python
#Creating a dlt pipeline with automated normalization
pipeline = dlt.pipeline(
    pipeline_name="ev_data",
    destination="duckdb",
    dataset_name="events_ds"
)

#run the pipeline
info = pipeline.run(paginated_getter(), table_name="events_table", write_disposition="replace")
print(info)

print(pipeline.dataset(dataset_type="default").schema.data_table_names())

pipeline.dataset(dataset_type="default").events_table.df()
#print(data.head())
```

```python
#Setting up data contracts with dlt
@dlt.resource(name="events", schema_contract={"tables":"evolve", "columns":"evolve", "data_type":"freeze"})
def paginated_getter():
    client = RESTClient(
        base_url = "https://api.github.com/repos/DataTalksClub/data-engineering-zoomcamp/events",
        #auth=BearerTokenAuth(token=API_TOKEN),
        paginator=HeaderLinkPaginator(links_next_key="next"),
    )

    for page in client.paginate(""):    #if endpoint isnt in the base url add it here
      for record in page:
        record['testing1']="abc"
        yield page

pipeline = dlt.pipeline(
    pipeline_name="ev_data_contract",
    destination="duckdb",
    dataset_name="events_ds"
)

info = pipeline.run(paginated_getter(), table_name="events_table", write_disposition="replace")
print(info)

pipeline.dataset(dataset_type="default").events_table.df()
```
