```python
plt.plot(epochs, train_loss, label = "Training loss") #x-axis, y-axis, label
#plt.show() in between shows both plots sperate
plt.plot(epochs, val_loss, label = "Validation loss") #put a second line graph on the same plot
plt.title("Loss v/s Epochs") #plot title
# plt.grid(True) adding grid lines
plt.xlabel("Epochs") #x-axis label
plt.ylabel("Loss") #y-axis-label
plt.legend() #adding a legend which has the labels
plt.show() #showing
```

```python
import pandas as pd
# Load the Titanic dataset
df = sns.load_dataset('titanic')
df.info()
```

```python
#initial data cleaning
# Drop less useful or high-missing columns
df.drop(columns=['deck', 'embark_town'], inplace=True)

# Fill missing age with median
df['age'].fillna(df['age'].median(), inplace=True)

# Fill embarked with mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

# Drop rows with any remaining missing values
df.dropna(inplace=True)

# Confirm cleanup
df.isnull().sum()
```

```python
#checking again
df.info()
```

```python
sns.countplot(x='pclass', data=df)
plt.title('Survival Count')
plt.xlabel('Pclass')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
sns.histplot(df['age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=df)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title('Survival by Gender')
plt.show()
```

```python
# Create a new feature: is_child
df['is_child'] = df['age'] < 18

# Create family size
df['family_size'] = df['sibsp'] + df['parch']
```
